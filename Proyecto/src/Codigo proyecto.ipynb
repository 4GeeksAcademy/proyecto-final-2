```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException, NoSuchElementException, ElementClickInterceptedException
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
#Archivo de Edad <25
# Configura el navegador (asegúrate de tener instalado el controlador correspondiente, como chromedriver)
driver = webdriver.Chrome()
driver.get('https://www.sofascore.com/tournament/football/turkey/trendyol-super-lig/52#id:42632')
# Acepta cookies si es necesario
try:
    cookie_button = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CLASS_NAME, 'fc-button.fc-cta-consent.fc-primary-button')))
    cookie_button.click()
    time.sleep(10)


    # Agrega clic en el botón de la pestaña detallados 
    button_detalles = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[1]/button[6]')))
    button_detalles.click()
    time.sleep(5)

    # agrega clic en el botón desmarcar la estadistica 
    button_stats1 = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[5]/div/div[3]/button[6]')))
    button_stats1.click()
    # agrega clic en el botón desmarcar la estadistica 
    button_stats2 = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[5]/div/div[3]/button[5]')))
    button_stats2.click()
    # agrega clic en el botón desmarcar la estadistica 
    button_stats3 = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[5]/div/div[3]/button[4]')))
    button_stats3.click()
    # agrega clic en el botón desmarcar la estadistica 
    button_stats4 = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[5]/div/div[3]/button[3]')))
    button_stats4.click()
    # agrega clic en el botón desmarcar la estadistica
    button_stats5 = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[5]/div/div[3]/button[2]')))
    button_stats5.click()
    # agrega clic en el botón desmarcar la estadistica 
    button_stats6 = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[5]/div/div[3]/button[1]')))
    button_stats6.click()
    time.sleep(10)

  
    # boton de edad
    button_selecedad = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[3]/div[1]/div/button')))
    button_selecedad.click()
    # Selecciona rango

    button_rango = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[3]/div[1]/div/div/div[1]/ul/li[4]')))
    button_rango.click()

    # Escribir la edad <25
    button_edad = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[3]/div[1]/input')))
    button_edad.click()
    button_edad.clear()
    button_edad.send_keys('25')
    time.sleep(10)
   
    # Botón de aplicar
    button_aplicar = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[6]/div/button[1]')))
    button_aplicar.click()

except:
    pass

# Inicializa una lista para almacenar los datos de todas las páginas
all_data = []

# Define la fila máxima que deseas alcanzar
fila_objetivo = 152

while True:
    try:
        # Espera a que la tabla se cargue completamente
        wait = WebDriverWait(driver, 20)
        wait.until(EC.presence_of_element_located((By.TAG_NAME, 'table')))
        
        # Obtén el HTML de la página después de cargar dinámicamente el contenido
        html_content = driver.page_source

        # Utiliza Beautiful Soup para encontrar la tabla en el HTML
        soup = BeautifulSoup(html_content, 'html.parser')
        table = soup.find('table')

        # Extrae los datos de la tabla y añádelos a la lista
        rows = table.find_all('tr')
        data = []
        for row in rows[1:]:
            row_data = [cell.text.strip() for cell in row.find_all(['th', 'td'])]
            data.append(row_data)
            
            # Verifica si se alcanzó la fila deseada
            if len(all_data) + len(data) >= fila_objetivo:
                break

        all_data.extend(data)

        # Intenta encontrar el botón de siguiente página
        try:
            next_button = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[7]/div/div/button[2]')

            # Hace clic en el botón de siguiente página usando JavaScript
            driver.execute_script("arguments[0].click();", next_button)

            # Espera a que la nueva página se cargue completamente
            wait.until(lambda driver: driver.execute_script('return document.readyState') == 'complete')

        except NoSuchElementException:
            break  # Sale del bucle si no hay más páginas

    except (TimeoutException, ElementClickInterceptedException):
        break  # Sale del bucle si se produce un tiempo de espera (timeout) o una excepción de clic interceptado

    # Verifica si se alcanzó la fila deseada después de cambiar de página
    if len(all_data) >= fila_objetivo:
        break

# Cierra el navegador fuera del bucle
driver.quit()

# Crea el DataFrame de pandas con todos los datos recopilados
headers = [header.text.strip() for header in rows[0].find_all(['th', 'td'])]
df_EDAD_M_25 = pd.DataFrame(all_data, columns=headers)
```

```python
#Archivo HABILIDAD APARICIONES
# Configura el navegador (asegúrate de tener instalado el controlador correspondiente, como chromedriver)
driver = webdriver.Chrome()
driver.get('https://www.sofascore.com/tournament/football/turkey/trendyol-super-lig/52#id:42632')
# Acepta cookies si es necesario
try:
    cookie_button = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.CLASS_NAME, 'fc-button.fc-cta-consent.fc-primary-button')))
    cookie_button.click()
    time.sleep(5)


    # Agrega clic en el botón de la pestaña detallados 
    button_detalles = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[1]/button[6]')))
    button_detalles.click()
    time.sleep(5)

    # agrega clic en el botón desmarcar la estadistica 
    button_stats1 = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[5]/div/div[3]/button[6]')))
    button_stats1.click()
    # agrega clic en el botón desmarcar la estadistica 
    button_stats2 = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[5]/div/div[3]/button[5]')))
    button_stats2.click()
    # agrega clic en el botón desmarcar la estadistica 
    button_stats3 = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[5]/div/div[3]/button[4]')))
    button_stats3.click()
    # agrega clic en el botón desmarcar la estadistica 
    button_stats4 = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[5]/div/div[3]/button[3]')))
    button_stats4.click()
    # agrega clic en el botón desmarcar la estadistica
    button_stats5 = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[5]/div/div[3]/button[2]')))
    button_stats5.click()
    # agrega clic en el botón desmarcar la estadistica 
    button_stats6 = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[5]/div/div[3]/button[1]')))
    button_stats6.click()
    time.sleep(10)


    # agrega clic en el botón  escoger tipo estadisticas ACUMULADO TODO
    
    
    # agrega clic en el botón desplegar estadisticas de interes
    button_selec = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[5]/div/div[1]/button[5]')))
    button_selec.click()
    time.sleep(10)

    # agrega clic en el botón de la pestaña detallados：
    button_hab1 = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[5]/div/div[2]/div/label[14]/div[1]')))
    button_hab1.click()
    time.sleep(10)
   
    # Botón de aplicar
    button_aplicar = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[6]/div/button[1]')))
    button_aplicar.click()

except:
    pass

# Inicializa una lista para almacenar los datos de todas las páginas
all_data = []

# Define la fila máxima que deseas alcanzar
fila_objetivo = 588

while True:
    try:
        # Espera a que la tabla se cargue completamente
        wait = WebDriverWait(driver, 20)
        wait.until(EC.presence_of_element_located((By.TAG_NAME, 'table')))
        
        # Obtén el HTML de la página después de cargar dinámicamente el contenido
        html_content = driver.page_source

        # Utiliza Beautiful Soup para encontrar la tabla en el HTML
        soup = BeautifulSoup(html_content, 'html.parser')
        table = soup.find('table')

        # Extrae los datos de la tabla y añádelos a la lista
        rows = table.find_all('tr')
        data = []
        for row in rows[1:]:
            row_data = [cell.text.strip() for cell in row.find_all(['th', 'td'])]
            data.append(row_data)
            
            # Verifica si se alcanzó la fila deseada
            if len(all_data) + len(data) >= fila_objetivo:
                break

        all_data.extend(data)

        # Intenta encontrar el botón de siguiente página
        try:
            next_button = driver.find_element(By.XPATH, '//*[@id="__next"]/main/div/div[3]/div/div[1]/div[1]/div[7]/div/div[7]/div/div/button[2]')

            # Hace clic en el botón de siguiente página usando JavaScript
            driver.execute_script("arguments[0].click();", next_button)

            # Espera a que la nueva página se cargue completamente
            wait.until(lambda driver: driver.execute_script('return document.readyState') == 'complete')

        except NoSuchElementException:
            break  # Sale del bucle si no hay más páginas

    except (TimeoutException, ElementClickInterceptedException):
        break  # Sale del bucle si se produce un tiempo de espera (timeout) o una excepción de clic interceptado

    # Verifica si se alcanzó la fila deseada después de cambiar de página
    if len(all_data) >= fila_objetivo:
        break

# Cierra el navegador fuera del bucle
driver.quit()

# Crea el DataFrame de pandas con todos los datos recopilados
headers = [header.text.strip() for header in rows[0].find_all(['th', 'td'])]
df_APARICIONES = pd.DataFrame(all_data, columns=headers)
```
